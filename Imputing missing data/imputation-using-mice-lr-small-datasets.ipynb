```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
LR.fit(X_train, Y_train)
```

```python
predict = LR.predict(X_test)
predict
```
